## 5. Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.
*   Normal: 18.5-24.9
*   Overweight: 25.0-29.9
*   Obesity: 30.0-34.9
*   Underweight: < 18.5


```python
#Membuat Function Kondisi Status BMI
def status_BMI(BMI):
  if BMI >= 18.5 and BMI <= 24.9:
    return 'Normal'
  elif BMI >= 25.0 and BMI <= 29.9:
    return 'Overweight'
  elif BMI >= 30.0 and BMI <= 34.9:
    return 'Obesity'
  elif BMI < 18.5:
    return 'Underweight'
#Menambahkan Kolom
df['Status BMI'] = df['BMI'].apply(status_BMI)
df.head()
```

# **SOAL 2**

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

```python
def calculate_investment(money_capital, tenure):
    # Cek apakah modal dan jangka waktu sesuai ketentuan
    if not (0 < money_capital <= 1000000000) or not (0 < tenure <= 25):
        print('Error: Modal atau jangka waktu tidak sesuai ketentuan.')
        return

    # Inisialisasi variabel akumulasi investasi
    total_investment = money_capital

    # Iterasi untuk setiap tahun
    for year in range(1, tenure + 1):
        # Hitung keuntungan dan akumulasi investasi
        profit = total_investment * 0.25
        total_investment += profit

        # Tampilkan hasil setiap tahun
        print(f'Year {year}, your total investment: {total_investment}')

# Menerima input dari pengguna
try:
    money_capital = int(input('>> Money Capital: '))
    tenure = int(input('>> Investment Tenure (years): '))
except ValueError:
    print('Error: Masukkan harus berupa bilangan bulat.')
    exit()

# Panggil fungsi untuk menghitung investasi
calculate_investment(money_capital, tenure)
```
